## 1. CRISP-DM

For this project, I will use CRISP-DM phylosophy. It stands for Cross-Industry Standard Process for Data Mining. It is a widely used process model for data mining, which provides a structured approach to planning and executing data mining projects. The process model is based on six phases, which are as follows:

1. **Business Understanding:** This phase involves understanding the goals and objectives of the data mining project, as well as defining the problem statement and success criteria.
2. **Data Understanding:** In this phase, data is collected and analyzed to gain a better understanding of the available data and to identify any quality issues or anomalies.
3. **Data Preparation:** This phase involves preparing the data for modeling, which includes cleaning, transforming, and selecting the relevant data.
4. **Modeling:** In this phase, various modeling techniques are applied to the prepared data to create the best possible model. EDA, Variable selection, Model selection using python or R.
5. **Evaluation:** The model created in the previous phase is evaluated against the success criteria defined in the first phase.
6. **Deployment:** The final phase involves deploying the model in the production environment and monitoring its performance.

CRISP-DM provides a standardized and structured approach to data mining projects, which helps to ensure that projects are completed on time, within budget, and with the desired results.

## 3.1.1 Univariate analysis of features 

* __Histograms__ and __bar charts__ are commonly used for univariate analysis. A histogram is a graphical representation of the distribution of a single variable. It shows the frequency distribution of the variable's values in the form of bars, where the height of each bar represents the frequency of occurrence of a value or range of values. A bar chart is similar to a histogram, but it is used to represent categorical data.
* __Boxplots__ are used for univariate analysis to visualize the distribution of a numerical variable. A boxplot displays the five-number summary of a dataset: minimum, first quartile, median, third quartile, and maximum. It can also show outliers beyond the whiskers. Boxplots are useful for identifying the presence of outliers, skewness, and the spread of the dataset.
* A __violin plot__ is used to visualize the distribution of numeric data across different levels or categories. It is similar to a box plot, but instead of only showing the summary statistics (minimum, maximum, median, quartiles) it also shows the density of the data at different values. One advantage of violin plots is that they allow us to compare the shapes of distributions across different groups or categories. For example, we can compare the salary distributions of different job titles or experience levels using a violin plot.

Other less known methods of univariate analysis are as follows:
* __Kernel Density Plots:__ Kernel density plots provide a smoothed estimate of the distribution of a numerical variable.
* __Boxen Plots:__ Boxen plots are similar to box plots but can show more details about the distribution, including additional quartiles and individual data points.
* __Probability Density Plots:__ Probability density plots depict the probability distribution of a continuous variable using a smooth curve.
* __Count Plots:__ Count plots are used to visualize the frequency of each category in a categorical variable

```python
# Define the keyword patterns for each general category
data_scientist_keywords = ['data scientist', 'data science']
data_engineer_keywords = ['data engineer', 'data infrastructure engineer', 'data devops engineer', 'cloud database engineer', 
                          'data operations engineer', 'etl developer', 'etl engineer']
data_analyst_keywords = ['data analyst', 'analytics engineer', 'data quality analyst', 'bi analyst', 'data analytics', 
                         'data visualization analyst', 'data operations analyst']
data_architect_keywords = ['data architect', 'bi developer']
data_manager_keywords = ['data manager', 'manager data management']
machine_learning_engineer_keywords = ['machine learning engineer', 'ml engineer', 'applied machine learning engineer',
                                      'machine learning software engineer','machine learning developer', 'mlops engineer',
                                      'machine learning infrastructure engineer', 'deep learning engineer']
machine_learning_scientist_keywords = ['machine learning scientist', 'machine learning research engineer', 
                                       'applied machine learning scientist', 'machine learning researcher',
                                       'machine learning manager']
computer_vision_engineer_keywords = ['computer vision']
nlp_engineer_keywords = ['nlp engineer']
research_scientist_keywords = ['research']
ai_scientist_keywords = ['ai']
applied_scientist_keywords = ['applied scientist']

# Function to classify job titles into general categories
def classify_job_title(job_title):
    job_title = job_title.lower()
    if any(keyword in job_title for keyword in data_scientist_keywords):
        return 'Data Scientist'
    elif any(keyword in job_title for keyword in data_engineer_keywords):
        return 'Data Engineer'
    elif any(keyword in job_title for keyword in data_analyst_keywords):
        return 'Data Analyst'
    elif any(keyword in job_title for keyword in data_architect_keywords):
        return 'Data Architect'
    elif any(keyword in job_title for keyword in data_manager_keywords):
        return 'Data Manager'
    elif any(keyword in job_title for keyword in machine_learning_engineer_keywords):
        return 'Machine Learning Engineer'
    elif any(keyword in job_title for keyword in machine_learning_scientist_keywords):
        return 'Machine Learning Scientist'
    elif any(keyword in job_title for keyword in computer_vision_engineer_keywords):
        return 'Computer Vision Engineer'
    elif any(keyword in job_title for keyword in nlp_engineer_keywords):
        return 'NLP Engineer'
    elif any(keyword in job_title for keyword in research_scientist_keywords):
        return 'Research Scientist'
    elif any(keyword in job_title for keyword in ai_scientist_keywords):
        return 'AI Scientist'
    elif any(keyword in job_title for keyword in applied_scientist_keywords):
        return 'Applied Scientist'
    else:
        return 'Undefined Category'

# Apply the job title classification to create a new 'general_title' column
salaries_usd_filtered['general_job_title'] = salaries_usd_filtered['job_title'].apply(classify_job_title)
```

## 3.1.2 Bivariate analysis of features 

Bivariate analysis is a statistical analysis method that involves analyzing the relationship between two variables. It is concerned with studying the behavior and interaction of two variables simultaneously to understand the relationship between them. Bivariate analysis is typically used to explore the association between two variables and to identify patterns, trends, or correlations between them. The analysis can be used to determine the strength, direction, and form of the relationship between two variables, and it can also be used to develop models to predict one variable based on the values of the other variable. Bivariate analysis is a fundamental component of data analysis and is often used as a starting point for more complex multivariate analysis. Following are some ways in which a bivariate analysis may be done.

* __Correlation analysis:__ A statistical technique that measures the strength and direction of the relationship between two continuous variables.
* __Heatmaps:__ Heatmaps can show the relationship between two variables by displaying the frequency or proportions in a tabular form using color intensities.
* __Clustered Bar Charts:__ Clustered bar charts are used to compare the frequency or proportions of different categories in two or more categorical variables.
* __Hexbin Plots:__ Hexbin plots are useful for visualizing the density of points in a scatter plot when there are a large number of overlapping points.
* __Violin Plots with Hue:__ Violin plots can be enhanced by adding a hue parameter to depict the distribution of a numeric variable across different categories
* __Scatter plot:__ A graphical representation of the relationship between two continuous variables, where each point represents a single observation.
* __Line chart:__ A chart that shows the relationship between two continuous variables over time or a range of values.
* __Regression analysis:__ A statistical technique used to model and analyze the relationship between two or more variables. It involves fitting a linear or non-linear equation to the data to predict the value of the dependent variable based on the values of the independent variables.
* __Contingency table analysis:__ A statistical technique used to analyze the relationship between two categorical variables by creating a table that shows the frequency of each combination of the variables.
* __Box plot:__ A graphical representation of the distribution of a continuous variable that shows the median, quartiles, and outliers.
* __Bubble chart:__ A type of chart that displays three dimensions of data by representing each data point with a circle, where the x-axis represents one variable, the y-axis represents another variable, and the size of the circle represents a third variable.

#### Converting `experience_level` feature into a numerical variable

The feature in the __salaries_usd_filtered__ called `experience_level` has four unique values. I want to convert these categorical values into numeric ones so they can be used in correlation heatmap and later for logistic regression. As they are levels so they can be converted into ordinal variable as follows:


| Category | Full Form | Numeric Value |
| -------- | --------- | ------------- |
| EN       | Entry-level | 0             |
| MI       | Junior Mid-level | 1         |
| SE       | Intermediate Senior-level | 2 |
| EX       | Expert Executive-level / Director | 3 |

Therefore, I will replace the values as below using the __OrdinalEncoder from SKLearn library__, I have already imported the encoder in the __imports__ section: 

```python
# Create an instance of OrdinalEncoder
ordinal_encoder = OrdinalEncoder(categories=[['EN', 'MI', 'SE', 'EX']])

# Reshape the column to be a 2-dimensional array
experience_level = df_heatmap['experience_level'].values.reshape(-1, 1)

# Encode the values
experience_level_encoded = ordinal_encoder.fit_transform(experience_level)

# Assign the encoded values back to the DataFrame
df_heatmap['experience_level_encoded'] = experience_level_encoded
```
